```python
ax=df['work_year'].plot(kind='kde', title="Year salary was paid")
ax.set_xlabel("Salary paid year")
```

```python
ax=df['employment_type'].value_counts().plot(kind='bar', title="Employment for the role")
ax.set_xlabel("Employment Type")
```

```python
ax=df['job_title'].value_counts().head(20).plot(kind='bar', title="Role worked during the year")
ax.set_xlabel("Job role")
```

```python
ax=df['salary'].value_counts().head(20).plot(kind='bar', title="Total salary paid")
ax.set_xlabel("Salary")
```

```python
ax=df['salary_currency'].value_counts().plot(kind='bar', title="Salary currency paid")
ax.set_xlabel("Salary Currency")
```

```python
ax=df['salary_in_usd'].plot(kind='kde', title="Salary in USD")
ax.set_xlabel("Salary in USD")
```

```python
ax=df['employee_residence'].value_counts().head(20).plot(kind='bar', title="Country of Residence")
ax.set_xlabel("Employees Country of Residence")
```

```python
ax=df['remote_ratio'].plot(kind='kde', title="Amount of work done remotly")
ax.set_xlabel("Work done Remotly")
```

```python
ax=df['company_location'].value_counts().head(20).plot(kind='bar', title="Number of People worked for the Company")
ax.set_xlabel("People worked for the Company")
```

```python
#Feature relatioship
sns.scatterplot(x="salary", y="work_year", hue="experience_level", data=df)
plt.show()
```

```python
sns.scatterplot(x='salary', y='work_year', hue='employment_type',data=df)
plt.show()
```

```python
# count frequency of each job title
job_counts= df['job_title'].value_counts()

# select top 10 job titles
top10_jobs= job_counts.nlargest(10).index.tolist()

# new dataframe with top 10 job titles
df_top10= df[df['job_title'].isin(top10_jobs)]

# scatter plot with job title
sns.scatterplot(x='salary',y='work_year',hue='job_title', data=df_top10)
ax.legend(fontsize='small')
plt.show()
```

```python
df_corr=df[['work_year','salary','salary_in_usd','remote_ratio']].corr()
df_corr
```
