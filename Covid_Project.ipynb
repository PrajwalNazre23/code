```python
data=data.apply("min")
```

```python
top_10_active = data.groupby(by= 'Active Cases').max()[['State/UT','Confirmed Cases']]
```

```python
sns.barplot(data=top_10_active.iloc[:10], y="State/UT",x="Confirmed Cases")
```

```python
top_10_active = data.groupby(by= 'Active Cases').max()[['State/UT','Confirmed Cases']]
sns.barplot(data=top_10_active.iloc[:10], y="State/UT",x="Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("State/UT")
plt.figure(figsize=(25,9))
plt.show()
```
